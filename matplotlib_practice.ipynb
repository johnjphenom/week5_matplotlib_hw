```python
# Import dependancies

import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Create the plot

plt.plot(x_axis, y_axis)
```

```python
# Create the plot with as.plt()

fig, ax = plt.subplots()

ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()

fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()

ax = plt.axes()

ax.plot(x_axis, y_axis)
```

```python
# Create the plot

plt.plot(x_axis, y_axis)

plt.show()
```

```python
# Create the plot and add a label for the legend.

plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')

# Create labels for the x and y axes

plt.xlabel("Date")

plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.

plt.ylim(0, 45)

#Create a title

plt.title("PyBer Fare by Month")

# Add a grid

plt.grid()

# Add a legend.

plt.legend()
```

```python
# Create the plot and add a label for the legend.

ax = plt.axes()
ax.plot(x_axis, y_axis, color ='green', label ='Boston', marker = "D" )

# Change the line color to green

ax.plot(linewidth=2)

ax.grid()

ax.set_xlabel('Date')

ax.set_ylabel('Fare($)')

ax.set_title("PyBer Fare by Month")

ax.set_ylim(5, 45)

ax.legend()
```

```python
# Set the x-axis to a list of strings for each month.

x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axix to a list of floats as the total fare in US dollars accumulated for each month.

y_axis = [10.2, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 28.90, 20.71, 20.09]
```

```python
plt.scatter(y_axis, x_axis, c= "red", label="Chicago")

plt.xlabel("Fare($)")

plt.ylabel("Date")

plt.title("PyBer Fare by Month")

plt.legend()

plt.gca().invert_yaxis()
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
y_axis_larger = []

for data in y_axis:
    
    y_axis_larger.append(data*3)
```

```python
#Skill Drill 5.1.7

fig, ax = plt.subplots()

ax.scatter(y_axis, x_axis, s = [i*5 for i in y_axis], c='skyblue', label = 'Boston', edgecolors = 'black', linewidth = 2, alpha = .8)

ax.set(xlabel=("Fare($)"), ylabel=("Date"), title = "PyBer Fare by Month")


ax.legend()

ax.invert_yaxis()

ax.set(xlim=(0, 50))

```

```python
plt.pie(y_axis, labels=x_axis)

plt.show()
```

```python

plt.subplots(figsize=(8, 8))

explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')



plt.show()
```

```python
# Assign 12 colors, one for each month.

colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

plt.subplots(figsize=(8, 8))

explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

plt.pie(y_axis, 
        
        explode=explode_values, 
        
        colors=colors,
        
        labels=x_axis, 
        autopct='%.1f%%')



plt.show()
```
